```python
# ============================
# Upload any file in Colab
# ============================
from google.colab import files
import pandas as pd

# Opens a "Choose File" button in Colab
uploaded = files.upload()

# Get filename (works for Excel/CSV)
for filename in uploaded.keys():
    file_path = filename
    print("Uploaded file:", file_path)

# Read Excel
if file_path.endswith(".xlsx") or file_path.endswith(".xls"):
    df = pd.read_excel(file_path, sheet_name="in")  # change sheet if needed
elif file_path.endswith(".csv"):
    df = pd.read_csv(file_path)
else:
    raise ValueError("Unsupported file format. Please upload .xlsx or .csv")

print("Shape:", df.shape)
df.head()
```

```python


# 1. Drop duplicates
df.drop_duplicates(inplace=True)

# 2. Drop insufficient columns
if "house_rules" in df.columns: df.drop(columns=["house_rules"], inplace=True)
if "license" in df.columns: df.drop(columns=["license"], inplace=True)

# 3. Clean price & service fee columns
for col in ["price", "service fee"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace("[$,]", "", regex=True).astype(float)

# 4. Handle missing values
df.dropna(inplace=True)

# 5. Fix spelling errors
if "neighbourhood group" in df.columns:
    df["neighbourhood group"] = df["neighbourhood group"].replace({"Brookln": "Brooklyn", "brooklyn": "Brooklyn"})

# 6. Fix data types
if "Construction year" in df.columns:
    df["Construction year"] = pd.to_numeric(df["Construction year"], errors="coerce")

# 7. Handle outliers (availability_365 should be between 0–365)
if "availability 365" in df.columns:
    df = df[(df["availability 365"] >= 0) & (df["availability 365"] <= 365)]

print("After Cleaning:", df.shape)
df.info()
```

```python
# ==========================
# Section 1: Import Libraries
# ==========================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from google.colab import files

# ==========================
# Section 2: Upload Excel from Google Drive/Local
# ==========================
uploaded = files.upload()  # choose your Excel file
file_name = list(uploaded.keys())[0]  # get uploaded file name
df = pd.read_excel(file_name)  # read Excel

# ==========================
# Section 2a: Normalize Column Names
# ==========================
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("Columns after normalization:\n", df.columns)

# ==========================
# Section 3: Quick Peek & Cleaning
# ==========================
print(df.head())
print(df.info())
print(df.describe())

# Drop duplicates
df = df.drop_duplicates()

# Fill missing values
df.fillna({'reviews_per_month': 0, 'name': 'unknown', 'host_name': 'unknown'}, inplace=True)

# Convert price column to numeric if exists
if 'price' in df.columns:
    df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# ==========================
# Section 4: Exploratory Data Analysis (EDA)
# ==========================
# Price Distribution
if 'price' in df.columns:
    plt.figure(figsize=(10,5))
    sns.histplot(df['price'], bins=50, kde=True)
    plt.title('Price Distribution')
    plt.xlabel('Price ($)')
    plt.ylabel('Count')
    plt.xlim(0, 1000)
    plt.show()

# Room Type Count
if 'room_type' in df.columns:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x='room_type', palette='Set2')
    plt.title('Room Type Distribution')
    plt.xlabel('Room Type')
    plt.ylabel('Count')
    plt.show()

# Top 10 Neighbourhoods by Listings
if 'neighbourhood' in df.columns:
    top_neigh = df['neighbourhood'].value_counts().nlargest(10)
    plt.figure(figsize=(10,5))
    sns.barplot(x=top_neigh.index, y=top_neigh.values, palette='Set3')
    plt.title('Top 10 Neighbourhoods by Listings')
    plt.xticks(rotation=45)
    plt.ylabel('Number of Listings')
    plt.show()

# Correlation Heatmap for numeric columns
numeric_cols = df.select_dtypes(include=np.number)
plt.figure(figsize=(12,8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# ==========================
# Section 5: Insights Examples
# ==========================
# Average Price per Room Type
if 'room_type' in df.columns and 'price' in df.columns:
    avg_price = df.groupby('room_type')['price'].mean().sort_values()
    print("Average Price per Room Type:\n", avg_price)

# Most Expensive Neighbourhoods
if 'neighbourhood' in df.columns and 'price' in df.columns:
    expensive_neigh = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)
    print("\nTop 10 Expensive Neighbourhoods:\n", expensive_neigh)

# Reviews vs Price Scatter Plot
if 'reviews_per_month' in df.columns and 'price' in df.columns:
    plt.figure(figsize=(10,6))
    sns.scatterplot(data=df, x='reviews_per_month', y='price', alpha=0.5)
    plt.title('Reviews per Month vs Price')
    plt.xlabel('Reviews per Month')
    plt.ylabel('Price ($)')
    plt.show()
```
